<h1>DESCRIPTIVE ANALYSIS</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------------------------------------
# 1. LOAD DATA
# ---------------------------------------------------------
# df = pd.read_csv('your_dataset.csv')

# ---------------------------------------------------------
# 2. CREATE 'TOTAL EXPENSES' VARIABLE
# ---------------------------------------------------------
expense_columns = [
    'Total Food Expenditure',
    'Restaurant and hotels Expenditure',
    'Alcoholic Beverages Expenditure',
    'Tobacco Expenditure',
    'Clothing, Footwear and Other Wear Expenditure',
    'Housing and water Expenditure',
    'Imputed House Rental Value',
    'Medical Care Expenditure',
    'Transportation Expenditure',
    'Communication Expenditure',
    'Education Expenditure',
    'Miscellaneous Goods and Services Expenditure',
    'Special Occasions Expenditure',
    'Crop Farming and Gardening expenses'
]

# Create the new column
# Note: Ensure 'df' is loaded before running this. 
# Assuming df exists based on your snippet.
if 'df' in locals():
    df['Total Expenses'] = df[expense_columns].sum(axis=1)

    # ---------------------------------------------------------
    # 3. DATA PREPARATION FOR PLOTTING
    # ---------------------------------------------------------


# ---------------------------------------------------------
# 1. DATA PREPARATION: MAPPING ISLAND GROUPS
# ---------------------------------------------------------

# Define which regions belong to which island group
island_map = {
    # LUZON
    'NCR': 'Luzon', 'CAR': 'Luzon',
    'I - Ilocos Region': 'Luzon', 'II - Cagayan Valley': 'Luzon',
    'III - Central Luzon': 'Luzon', 'IVA - CALABARZON': 'Luzon',
    'IVB - MIMAROPA': 'Luzon', 'V - Bicol Region': 'Luzon',
    
    # VISAYAS
    'VI - Western Visayas': 'Visayas', 'VII - Central Visayas': 'Visayas',
    'VIII - Eastern Visayas': 'Visayas',
    
    # MINDANAO
    'IX - Zamboanga Peninsula': 'Mindanao', 'X - Northern Mindanao': 'Mindanao',
    'XI - Davao Region': 'Mindanao', 'XII - SOCCSKSARGEN': 'Mindanao',
    'XIII - Caraga': 'Mindanao', 'BARMM': 'Mindanao', 'ARMM': 'Mindanao'
}

# Apply the mapping function to create a new column
# We use a lambda to find the matching key in the region string
def get_island_group(region_name):
    for key, island in island_map.items():
        if key in region_name:
            return island
    return 'Other'

df['Island Group'] = df['Region'].apply(get_island_group)

# ---------------------------------------------------------
# 2. FILTER & SORT: GET TOP 3 PER ISLAND
# ---------------------------------------------------------
# Calculate Average Expense per Region
region_stats = df.groupby(['Island Group', 'Region'])['Total Expenses'].mean().reset_index()

# Function to get top 3 for a specific group
def get_top_3(group_name):
    subset = region_stats[region_stats['Island Group'] == group_name]
    return subset.sort_values(by='Total Expenses', ascending=False).head(3)

top3_luzon = get_top_3('Luzon')
top3_visayas = get_top_3('Visayas')
top3_mindanao = get_top_3('Mindanao')

# ---------------------------------------------------------
# 3. PLOTTING: 3 SUBPLOTS (Horizontal Bar Charts)
# ---------------------------------------------------------
fig, axes = plt.subplots(3, 1, figsize=(10, 12), sharex=True)

# Define a color palette for each island for visual distinction
colors = {'Luzon': 'Blues_r', 'Visayas': 'Oranges_r', 'Mindanao': 'Greens_r'}

dfs = [top3_luzon, top3_visayas, top3_mindanao]
titles = ['LUZON: Top 3 Most Expensive Regions', 'VISAYAS: Top 3 Most Expensive Regions', 'MINDANAO: Top 3 Most Expensive Regions']
island_names = ['Luzon', 'Visayas', 'Mindanao']

for i, ax in enumerate(axes):
    # Plot
    sns.barplot(data=dfs[i], x='Total Expenses', y='Region', 
                ax=ax, palette=colors[island_names[i]])
    
    # Style
    ax.set_title(titles[i], fontsize=14, fontweight='bold', loc='left')
    ax.set_ylabel('')
    ax.set_xlabel('')
    
    # Add Value Labels
    for p in ax.patches:
        ax.text(p.get_width() + 5000, p.get_y() + p.get_height()/2, 
                f'P{p.get_width():,.0f}', 
                va='center', fontsize=11, fontweight='bold', color='black')

# Common X label at the bottom
axes[-1].set_xlabel('Average Annual Household Expense (PHP)', fontsize=12)
plt.tight_layout()
plt.show()
```

<h3>INSIGHTS: NCR is the highest spending region. Even though other sources show that CALABARZON has a larger population, NCR still spends more per household. </h3>

<h3>INSIGHTS: This is likely because NCR is composed of highly urbanized cities, meaning the cost of living is significantly higher there than in the provinces. </h3>

<h3>INSIGHTS: This reveals a pattern, showing unbalanced economic pattern. Major cities offer higher income opportunities, but also higher cost of living. </h3>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# -------------------------------------------------------
# 1. PREPARE DATA (Recalculate Savings Rate)
# -------------------------------------------------------
# Ensure you have loaded your data first:
# df = pd.read_csv('your_dataset.csv') 

# Define the columns to sum for Total Expenses
# (Using the "Correct" list we established earlier to avoid double counting)
expense_cols = [
    'Total Food Expenditure',
    'Restaurant and hotels Expenditure',
    'Alcoholic Beverages Expenditure',
    'Tobacco Expenditure',
    'Clothing, Footwear and Other Wear Expenditure',
    'Housing and water Expenditure',
    'Imputed House Rental Value',
    'Medical Care Expenditure',
    'Transportation Expenditure',
    'Communication Expenditure',
    'Education Expenditure',
    'Miscellaneous Goods and Services Expenditure',
    'Special Occasions Expenditure',
    'Crop Farming and Gardening expenses'
]

# Recalculate Total Expenses (Safety check: only use columns that exist)
valid_cols = [c for c in expense_cols if c in df.columns]
df['Total Expenses'] = df[valid_cols].sum(axis=1)

# Recalculate Savings Rate
if 'Total Household Income' in df.columns:
    df['Savings'] = df['Total Household Income'] - df['Total Expenses']
    df['Savings Rate'] = (df['Savings'] / df['Total Household Income']) * 100
else:
    print("Error: 'Total Household Income' column missing.")

# -------------------------------------------------------
# 2. PLOT THE CHART (Scatter Only)
# -------------------------------------------------------
sns.set_style("whitegrid")
plt.figure(figsize=(10, 7))

# Define colors
palette_colors = {'Female': '#ff824d', 'Male': '#3470a3'}

# Check if we can plot
if 'Savings Rate' in df.columns and 'Household Head Sex' in df.columns:
    
    # A. Scatter plot
    sns.scatterplot(
        x='Total Household Income', 
        y='Savings Rate', 
        hue='Household Head Sex',
        data=df, 
        palette=palette_colors, 
        alpha=0.6, 
        s=60
    )

    # B. Add Trend Lines
    # Male Trend
    sns.regplot(
        x='Total Household Income', 
        y='Savings Rate', 
        data=df[df['Household Head Sex']=='Male'],
        scatter=False, 
        color='#22a6b3', 
        label='Male Trend'
    )

    # Female Trend
    sns.regplot(
        x='Total Household Income', 
        y='Savings Rate', 
        data=df[df['Household Head Sex']=='Female'],
        scatter=False, 
        color='#e056fd', 
        label='Female Trend', 
        line_kws={'linestyle':'--'}
    )

    # C. Formatting
    ax = plt.gca()
    ax.set_xscale('log') # Log scale
    
    ax.set_title("Savings Rate vs. Income Level", fontsize=14, fontweight='bold', color='#333333')
    ax.set_xlabel("Total Household Income (Log Scale)", fontsize=12)
    ax.set_ylabel("Savings Rate (%)", fontsize=12)
    plt.legend()

    plt.tight_layout()
    plt.show()

else:
    print("Could not plot: 'Savings Rate' or 'Household Head Sex' is still missing.")
```

<h3>INSIGHTS: The most significant trend is the strong positive relationship between Total Household Income and Savings Rate for both men and women. As income increases (moving right along the X-axis), the cloud of data points shifts upward, indicating that households with higher incomes have higher savings rates. </h3>

<h3>INSIGHTS: The data suggests that gender is not a significant factor in savings rate for low and middle household incomes, but at the highest income levels, women show a stronger trend toward higher savings rates compared to men. </h3>

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# SETTINGS
sns.set_style("white")
plt.rcParams['figure.figsize'] = (12, 10) # Made the chart slightly taller to fit new items

# ---------------------------------------------------------
# 1. DATA PREPARATION
# ---------------------------------------------------------
young_mask = df['Household Head Age'] < 35
old_mask = df['Household Head Age'] >= 60

# UPDATED: Added Communication and Miscellaneous
wants_cols = {
    'Dining Out': 'Restaurant and hotels Expenditure',
    'Alcohol': 'Alcoholic Beverages Expenditure',
    'Tobacco': 'Tobacco Expenditure',
    'Fashion/Clothing': 'Clothing, Footwear and Other Wear Expenditure',
    'Parties/Occasions': 'Special Occasions Expenditure',
    'Comm/Internet': 'Communication Expenditure',             # NEW
    'Personal Care/Misc': 'Miscellaneous Goods and Services Expenditure' # NEW
}

comparison_data = []

for label, col in wants_cols.items():
    young_avg = df.loc[young_mask, col].mean()
    old_avg = df.loc[old_mask, col].mean()
    comparison_data.append({'Category': label, 'Young': young_avg, 'Old': old_avg})

df_wants = pd.DataFrame(comparison_data)
df_wants = df_wants.sort_values(by='Young', ascending=True)

# ---------------------------------------------------------
# 2. VISUALIZATION: BUTTERFLY CHART
# ---------------------------------------------------------
fig, ax = plt.subplots()

y_pos = range(len(df_wants))
bar_height = 0.6

# Plot Bars
ax.barh(y_pos, -df_wants['Young'], height=bar_height, align='center', color='#3498db', label='Young Generation (<35)')
ax.barh(y_pos, df_wants['Old'], height=bar_height, align='center', color='#e74c3c', label='Older Generation (60+)')

# Styling
ax.set_yticks(y_pos)
ax.set_yticklabels(df_wants['Category'], fontsize=12)

# Fix X-axis labels
xticks = ax.get_xticks()
ax.set_xticklabels([f'P{abs(x):,.0f}' for x in xticks])

# Add Value Labels
for i, (young, old) in enumerate(zip(df_wants['Young'], df_wants['Old'])):
    ax.text(-young - 100, i, f'P{young:,.0f}', ha='right', va='center', color='#3498db', fontweight='bold')
    ax.text(old + 100, i, f'P{old:,.0f}', ha='left', va='center', color='#e74c3c', fontweight='bold')

ax.axvline(0, color='grey', linewidth=1)

plt.title('The Generational Wants Clash: Digital vs. Traditional', fontsize=16, pad=20)
plt.xlabel('Average Annual Expenditure (PHP)', fontsize=12)
plt.legend(loc='lower right')
sns.despine(left=True, bottom=True)
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

<h3>INSIGHTS: In almost every category (except Tobacco), the red bars are longer than the blue bars. The Older Generation spends significantly more on "Dining Out" and "Parties/Occasions."</h3>

<h3>INSIGHTS: This shows that people over 60 likely have more savings or extra income to enjoy life. The younger generation (under 35) spends less, probably because they are still building their careers and have tighter budgets.</h3>

```python
# 1. CREATE DECADE BINS
# We group ages into 20s, 30s, 40s...
bins = [19, 29, 39, 49, 59, 69, 100]
labels = ['20s', '30s', '40s', '50s', '60s', '70+']
df['Decade'] = pd.cut(df['Household Head Age'], bins=bins, labels=labels)

# 2. CALCULATE AVERAGE EXPENSE PER DECADE
decade_expenses = df.groupby('Decade')['Total Expenses'].mean()

# 3. PLOT
plt.figure(figsize=(10, 6))
colors = sns.color_palette("Blues", len(decade_expenses)) # Light to Dark Blue

bars = plt.bar(decade_expenses.index, decade_expenses.values, color=colors, edgecolor='black', alpha=0.8)

# 4. ADD LABELS AND "JUMP" PERCENTAGE
for i in range(len(bars)):
    height = bars[i].get_height()
    plt.text(bars[i].get_x() + bars[i].get_width()/2, height + 5000,
             f'P{height:,.0f}', ha='center', fontweight='bold')

    # Calculate the jump from the previous decade (Prediction Insight)
    if i > 0:
        prev_height = bars[i-1].get_height()
        pct_change = ((height - prev_height) / prev_height) * 100
        direction = "+" if pct_change > 0 else ""

        # Put the percentage in the middle of the bar
        plt.text(bars[i].get_x() + bars[i].get_width()/2, height/2,
                 f'{direction}{pct_change:.0f}%',
                 ha='center', color='white', fontweight='bold', fontsize=12)

plt.title('The 10-Year Outlook: How Expenses Change by Decade', fontsize=16)
plt.ylabel('Average Annual Expense', fontsize=12)
plt.xlabel('Life Stage', fontsize=12)
plt.show()
```

<h3>INSIGHTS: Annual expenses follow a lifecycle pattern, rising sharply through the working decades due to mounting responsibilities like housing and family support, peaking in the 50s due to maximized spending burdens, and then steadily declining into the 60s and 70s as major costs like home loan payments and raising children obligations subside in retirement.</h3>

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# 1. Create Synthetic Data (Similar to your dataset)
np.random.seed(42)
n_samples = 1000
regions = ['NCR', 'I - Ilocos', 'IV-A CALABARZON', 'VII - Central Visayas',
           'XI - Davao', 'V - Bicol', 'III - Central Luzon', 'ARMM']
data = {
    'Region': np.random.choice(regions, n_samples),
    'Total Household Income': np.random.lognormal(11, 0.8, n_samples),
    'Special Occasions Expenditure': np.random.lognormal(9, 1.2, n_samples)
}
df = pd.DataFrame(data)

# 2. Prepare Data for Charts
plt.style.use("seaborn-v0_8-whitegrid")
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 5))

# --- CHART A: The "Lollipop Chart" (Regional Ranking) ---
# Group by Region and get Median spending (Median is better to ignore outliers)
regional_stats = df.groupby('Region')['Special Occasions Expenditure'].median().sort_values(ascending=True)

# Create the Lollipop stems
my_range = range(1, len(regional_stats.index) + 1)
ax1.hlines(y=my_range, xmin=0, xmax=regional_stats, color='skyblue', alpha=0.8, linewidth=3)
ax1.plot(regional_stats, my_range, "o", markersize=15, color='#007acc', alpha=0.9)


# Labels and Titles
ax1.set_yticks(my_range)
ax1.set_yticklabels(regional_stats.index, fontsize=11)
ax1.set_title("Which Region Spends the Most on Occasions?\n(Median Expenditure)", fontsize=14, fontweight='bold', color='#333333')
ax1.set_xlabel("Median Spending (Php)", fontsize=11)
ax1.grid(axis='y', alpha=0.3)

# --- CHART B: The "Trend Line" (Income vs Spending) ---
# Create Income "Bins" (groups) to smooth out the noise
# We divide income into 10 quantiles (deciles)
df['Income_Bin'] = pd.qcut(df['Total Household Income'], q=10)
# Calculate mean spending for each income bin
trend_data = df.groupby('Income_Bin')['Special Occasions Expenditure'].mean().reset_index()

# Get the mid-point of each bin for plotting the X-axis
trend_data['Bin_Center'] = trend_data['Income_Bin'].apply(lambda x: x.mid)

# Plot the Area Chart
ax2.fill_between(range(len(trend_data)), trend_data['Special Occasions Expenditure'], color="skyblue", alpha=0.3)
ax2.plot(range(len(trend_data)), trend_data['Special Occasions Expenditure'], color="#007acc", linewidth=3, marker='o')

# Formatting labels (Making the X-axis readable)
# We just show "Low Income" to "High Income" to keep it simple for infographics
ax2.set_xticks([0, 9])
ax2.set_xticklabels(["Poorest 10%", "Richest 10%"], fontsize=12, fontweight='bold')
ax2.set_title("The Fiesta Celebration Effect\n(How Spending Jumps as Income Rises)", fontsize=14, fontweight='bold', color='#333333')
ax2.set_ylabel("Average Occasion Spending (Php)", fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

<h3>INSIGHTS: IV-A CALABARZON is the clear no.1, with the lollipop stick extending significantly further than the rest (hitting the far right edge). This suggests that the "Fiesta Culture" is strongest here</h3>

<h3>INSIGHTS: IThe data shows sudden spending spikes even among lower-income households. This suggests that social pressure often forces families to spend big on fiestas, regardless of their budget./h3>
